```python
# =========================================================
# EXPLORATORY DATA ANALYSIS (EDA) PROJECT
# Internship Project using Python
# =========================================================

# ---------------------------------------------------------
# STEP 1 : IMPORT REQUIRED LIBRARIES
# ---------------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------------
# STEP 2 : CREATE SAMPLE DATASET
# ---------------------------------------------------------

np.random.seed(101)

data = {
    "Customer_ID": np.arange(1, 501),
    "Age": np.random.randint(18, 65, 500),
    "Gender": np.random.choice(["Male", "Female"], 500),
    "City": np.random.choice(
        ["Hyderabad", "Chennai", "Bangalore", "Mumbai", "Delhi"],
        500
    ),
    "Product_Category": np.random.choice(
        ["Electronics", "Fashion", "Groceries", "Sports"],
        500
    ),
    "Purchase_Amount": np.random.randint(500, 50000, 500),
    "Rating": np.random.randint(1, 6, 500)
}

df = pd.DataFrame(data)

print("========== DATASET ==========")
print(df.head())

# ---------------------------------------------------------
# STEP 3 : BASIC INFORMATION
# ---------------------------------------------------------

print("\n========== BASIC INFORMATION ==========")

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

# ---------------------------------------------------------
# STEP 4 : CHECK MISSING VALUES
# ---------------------------------------------------------

print("\n========== MISSING VALUES ==========")

print(df.isnull().sum())

# ---------------------------------------------------------
# STEP 5 : STATISTICAL SUMMARY
# ---------------------------------------------------------

print("\n========== STATISTICAL SUMMARY ==========")

print(df.describe())

# ---------------------------------------------------------
# STEP 6 : UNIVARIATE ANALYSIS
# ---------------------------------------------------------

sns.set_style("whitegrid")

# Distribution of Purchase Amount
plt.figure(figsize=(10, 6))

sns.histplot(
    df["Purchase_Amount"],
    bins=30,
    kde=True
)

plt.title("Distribution of Purchase Amount")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")
plt.show()

# ---------------------------------------------------------
# STEP 7 : PRODUCT CATEGORY ANALYSIS
# ---------------------------------------------------------

plt.figure(figsize=(10, 6))

sns.countplot(
    x="Product_Category",
    data=df
)

plt.title("Count of Product Categories")
plt.xlabel("Product Category")
plt.ylabel("Count")
plt.show()

# ---------------------------------------------------------
# STEP 8 : CITY WISE SALES ANALYSIS
# ---------------------------------------------------------

city_sales = df.groupby("City")["Purchase_Amount"].sum()

plt.figure(figsize=(8, 8))

plt.pie(
    city_sales,
    labels=city_sales.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("City Wise Sales Contribution")
plt.show()

# ---------------------------------------------------------
# STEP 9 : GENDER VS PURCHASE ANALYSIS
# ---------------------------------------------------------

plt.figure(figsize=(8, 6))

sns.boxplot(
    x="Gender",
    y="Purchase_Amount",
    data=df
)

plt.title("Gender vs Purchase Amount")
plt.show()

# ---------------------------------------------------------
# STEP 10 : AGE VS PURCHASE ANALYSIS
# ---------------------------------------------------------

plt.figure(figsize=(10, 6))

sns.scatterplot(
    x="Age",
    y="Purchase_Amount",
    hue="Gender",
    data=df
)

plt.title("Age vs Purchase Amount")
plt.xlabel("Age")
plt.ylabel("Purchase Amount")
plt.show()

# ---------------------------------------------------------
# STEP 11 : CORRELATION ANALYSIS
# ---------------------------------------------------------

print("\n========== CORRELATION MATRIX ==========")

numeric_df = df.select_dtypes(include=np.number)

correlation = numeric_df.corr()

print(correlation)

plt.figure(figsize=(8, 6))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")
plt.show()

# ---------------------------------------------------------
# STEP 12 : CATEGORY-WISE AVERAGE PURCHASE
# ---------------------------------------------------------

avg_purchase = df.groupby(
    "Product_Category"
)["Purchase_Amount"].mean()

plt.figure(figsize=(10, 6))

sns.barplot(
    x=avg_purchase.index,
    y=avg_purchase.values
)

plt.title("Average Purchase by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Average Purchase Amount")
plt.show()

# ---------------------------------------------------------
# STEP 13 : TOP INSIGHTS
# ---------------------------------------------------------

print("\n========== KEY INSIGHTS ==========")

highest_category = avg_purchase.idxmax()

highest_city = city_sales.idxmax()

highest_purchase = df["Purchase_Amount"].max()

print(f"Highest Spending Product Category : {highest_category}")

print(f"Top Revenue Generating City : {highest_city}")

print(f"Highest Single Purchase Amount : ₹{highest_purchase}")

print("\nEDA Project Completed Successfully!")

# =========================================================
# END OF PROJECT
# =========================================================
```
